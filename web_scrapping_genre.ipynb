### Webscrapping from everynoise.com

Aqui encontramos todas las variedades de techno y techhouse europeas, dont stop til you get enough 

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
import os 
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
import re
```

```python
url2 = "http://everynoise.com/countries.html"
```

```python
response = requests.get(url2)
```

```python
rank_genres = BeautifulSoup(response.content)
```

```python
genres_names = rank_genres.find_all(
    name="td", 
    class_="column section"
)
```

```python
 genres_names[0].find_all('a')
```

```python
eu_us_genres = []
for row in genres_names:
    eu_us_genres.append([cell.text for cell in row.find_all() if cell.text])

```

```python
tech_house = [a for region in eu_us_genres for a in region if re.match(".*tech house.*", a)]
```

```python
techno = [a for region in eu_us_genres for a in region if re.match(".*techno.*", a)]
```

```python
techdf = pd.DataFrame(techno)
```

```python
housedf = pd.DataFrame(tech_house)
```

```python
dance_energenres = pd.concat([techdf,housedf], ignore_index=True, axis=1)
```

```python
dance_energenres.columns = ['techno', 'techhouse']
```

```python
dance_energenres.to_csv("web_scrapping_genre.csv", index=False)
```
